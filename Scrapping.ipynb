```python
import time
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.common.action_chains import ActionChains
import pandas as pd
import numpy as np
```

```python
d = webdriver.Chrome(executable_path='chromedriver')
d.implicitly_wait(10)
```

```python
d.get('https://seller.flipkart.com/')
```

```python
def act(url):
    d.get(url)
    for i in range(10,51,10):
        ActionChains(d).move_to_element(d.find_element_by_xpath('//li[contains(@class,"cf-list'+str(i)+'")]')).perform()
    l1 = d.find_elements_by_xpath('//div/small[@class="txt-reference-id"][2]')
    R_ids = [i.text for i in l1]
    return R_ids
```

```python
def od_reas(R_ids,End_str):
    data = []
    for r in R_ids:
        d.get("https://seller.flipkart.com/index.html#dashboard/viewIssueManagementTicket/"+r+End_str)
        time.sleep(4)
        d.find_elements_by_xpath('//i[contains(@class,"expand")]')[-1].click()
        time.sleep(2)
        t = d.find_elements_by_xpath('//div[contains(@class,"thread-text")]')
        s = d.find_elements_by_tag_name('td')
        Od_Id = 's'
        for j in range(len(s)):
            if s[j].text =='Order ID:':
                Od_Id = s[j+1].text
                break
        data.append([Od_Id,t[-1].text])
        print(Od_Id)
    return data
```

```python
R_ids1 = act('https://seller.flipkart.com/index.html#dashboard/sellerSupport/manageTickets/spf/tickets')
data1 = od_reas(R_ids1,'/SPF')
```

```python
df1 = pd.DataFrame(data1,columns = ['OrderId','Reason'])
df1
```

```python
d.find_elements_by_xpath('//div[contains(@class,"pull-right")]')
```

```python
R_ids2 = act('https://seller.flipkart.com/index.html#dashboard/sellerSupport/manageTickets/general/tickets')
t = d.find_elements_by_xpath('//span[contains(@class,"pull-right")]')
Rs = []
for i in range(len(t)):
    if 'Claim' in t[i].text:
        Rs.append(R_ids2[i])
print(Rs)
data2 = od_reas(Rs,'/SELLER_SELF_SERVE')
```

```python
for entry in data2:
    if entry[0].startswith('OD') and entry[0] not in df1.OrderId:
        df1.loc[len(df1.index)] = entry
```

```python
df1.drop(0,axis=0,inplace=True)
```

```python
df1.reset_index(drop=True,inplace=True)
```
